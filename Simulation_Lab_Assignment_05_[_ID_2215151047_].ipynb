Apply the One-Sample Kolmogorov-Smirnov (K-S) Test on a dataset of daily temperatures to determine if the sample data comes from a specified normal distribution.

```python
#Importing the libraries

import numpy as np
from scipy.stats import kstest, norm
import matplotlib.pyplot as plt
```

```python
# Creating a dataset using the temperature data
temp_data = np.array([
    [15],
    [17],
    [16],
    [14],
    [18],
    [20],
    [19],
    [15],
    [21],
    [22]
])
```

```python
# Extracting the data
tem_data = temp_data[:, 0]
print("Temperatues of the last 10 days", tem_data)
```

```python
# Defining mean and standard deviation for the normal distribution
mean_val = np.mean(temp_data)
std_dev  = np.std(temp_data, ddof=0)
print(f"Assumed Normal Distribution: μ = {mean_val:.2f}, σ = {std_dev:.4f}")
```

```python
# Center and scale the data for standard normal comparison
scaled_data = (tem_data - mean_val) / std_dev #standardization
```

```python
# Performing the one-sample Kolmogorov-Smirnov test
ks_result = kstest(scaled_data, 'norm')
```

```python
# Displaying the test results
print(f"Test decision (h): {int(ks_result.pvalue < 0.05)}")
print(f"p-value: {ks_result.pvalue:.4f}")
print(f"KS statistic: {ks_result.statistic:.4f}")
```

```python
# Computing empirical CDF
sorted_data = np.sort(scaled_data)
ecdf_y = np.arange(1, len(scaled_data)+1) / len(scaled_data)

# Computing standard normal CDF
normal_cdf = norm.cdf(sorted_data)
```

```python
# Plotting the empirical CDF vs standard normal CDF
plt.figure(figsize=(8, 5))
plt.plot(sorted_data, ecdf_y, label='Empirical CDF', color='blue', linewidth=2)
plt.plot(sorted_data, normal_cdf, label='Standard Normal CDF', color='red', linestyle='--', linewidth=2)
plt.title("K-S Test: Empirical vs Normal CDF")
plt.xlabel("Scaled Exam Scores")
plt.ylabel("Cumulative Probability")
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```

**Determining the Result and Conclusion**

From the result, the null hypothesis can't be rejected. The p_value of KS_Test is greater than the significance level. The temperatures are not showing any sign of being extreme outliers. So, in this case, the null hypothesis is correct.
