```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Task 1

vec1 = np.random.rand(15)
vec2 = np.random.rand(15)

# Plotting

plt.figure(figsize=(8, 6))
plt.plot(vec1, marker='o', linestyle='-', label='Vector 1')
plt.plot(vec2, marker='s', linestyle='--', label='Vector 2')
plt.xlabel('Index')
plt.ylabel('Random Value')
plt.title('Two Vectors of 15 Random Floats')
plt.legend()
plt.show()
```

```python
#Task 2

matrix = np.random.rand(4, 4)
print(matrix)

# Plotting the heatmap

sns.heatmap(matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.xlabel("Columns")
plt.ylabel("Rows")
plt.title('Heatmap of 3x3 Matrix')
plt.show()
```

```python
# Task 3

A = np.random.randint(0, 10, size=(4, 4))
B = np.random.randint(0, 10, size=(4, 4))

print("Matrix A:\n", A)
print("Matrix B:\n", B)

sum = A + B
sub = A - B
mul = np.dot(A, B)

sum_flat  = sum.flatten()
diff_flat = sub.flatten()
prod_flat = mul.flatten()


fig, axes = plt.subplots(1, 3, figsize=(15, 5))


axes[0].bar(range(1, 17), sum.flatten(), color='b')
axes[0].set_title('Matrix Addition')
axes[0].set_xlabel('Index')
axes[0].set_ylabel('Value')

# Difference
axes[1].bar(range(1, 17), sub.flatten(), color='g')
axes[1].set_title('Matrix Subtraction')
axes[1].set_xlabel('Index')
axes[1].set_ylabel('Value')

# Product
axes[2].bar(range(1, 17), mul.flatten(), color='r')
axes[2].set_title('Matrix Multiplication')
axes[2].set_xlabel('Index')
axes[2].set_ylabel('Value')

plt.tight_layout()
plt.show()
```
