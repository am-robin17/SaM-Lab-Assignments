— Calculate the t-statistic and p-value.

— Interpret and visualize the result at a 0.05 significance levelimport numpy as np





```python
import numpy as np
import scipy.stats as stats #new library
import matplotlib.pyplot as plt
```

```python
#Defining the dataset
data = np.array([5.2, 5.4, 5.1, 5.6, 5.3, 5.8])

#Defining the population mean (hypothesized mean)
population_mean = 5.5  # Null Hypothesis (H₀): The average is 5.5

#Performing the one-sample t-test
t_statistic, p_value = stats.ttest_1samp(data, population_mean)

```

```python
#Conclusion
alpha = 0.05  # Significance level
if p_value < alpha:
    print(f'Reject the null hypothesis. The average class score is significantly different from {population_mean:.2f}.')
else:
    print(f'Fail to reject the null hypothesis. The average class score is not significantly different from {population_mean:.2f}.')
```

```python
#Visualizing the data
plt.figure(figsize=(8,6))
plt.bar(1, np.mean(data), color="springgreen", label='Sample Mean')
plt.errorbar(1, np.mean(data), yerr=np.std(data), fmt='k', linewidth=1.5, label='Standard Deviation')
plt.axhline(y=population_mean, color='blue', linestyle='--', linewidth=1.5, label='Hypothesized Mean')
plt.title('Sample Mean vs Hypothesized Mean')
plt.xlim(0.5,1.5)
plt.xlabel('Sample')
plt.ylabel('Mean Value')
plt.legend()
plt.grid(True)
plt.show()
```
