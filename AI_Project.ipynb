```python
import random

# Sample anime/movies and their genres
anime_database = {
    "Death Note": ["Thriller", "Mystery", "Psychological"],
    "Attack on Titan": ["Action", "Fantasy", "Horror"],
    "Spirited Away": ["Fantasy", "Adventure"],
    "Your Name": ["Romance", "Drama"],
    "Naruto": ["Action", "Adventure"],
    "One Piece": ["Action", "Adventure", "Fantasy"],
    "Fullmetal Alchemist: Brotherhood": ["Action", "Adventure", "Fantasy"],
    "Tokyo Ghoul": ["Action", "Horror", "Psychological"],
    "Dragon Ball Z": ["Action", "Adventure"],
    "One Punch Man": ["Action", "Comedy", "Superhero"],
    "My Hero Academia": ["Action", "Superhero"],
    "Cowboy Bebop": ["Action", "Sci-Fi"],
    "Neon Genesis Evangelion": ["Action", "Mecha", "Psychological"],
    "Hunter x Hunter": ["Action", "Adventure", "Fantasy"],
    "Steins;Gate": ["Sci-Fi", "Thriller"],
    "Code Geass": ["Action", "Mecha", "Psychological"],
    "Fairy Tail": ["Action", "Adventure", "Fantasy"],
    "Demon Slayer": ["Action", "Fantasy", "Supernatural"],
    "Black Clover": ["Action", "Adventure", "Fantasy"],
    "JoJo's Bizarre Adventure": ["Action", "Adventure", "Supernatural"],
    # 50 additional anime added to the database
    "Bleach": ["Action", "Adventure", "Supernatural"],
    "Dragon Ball": ["Action", "Adventure"],
    "My Neighbor Totoro": ["Fantasy", "Adventure"],
    "Princess Mononoke": ["Fantasy", "Adventure"],
    "Grave of the Fireflies": ["Drama", "War"],
    "The Promised Neverland": ["Mystery", "Psychological", "Thriller"],
    "Death Parade": ["Psychological", "Supernatural", "Thriller"],
    "Attack on Titan: Junior High": ["Comedy", "Parody"],
    "One Piece: Stampede": ["Action", "Adventure", "Comedy"],
    "Dragon Ball Super": ["Action", "Adventure"],
    "Naruto Shippuden": ["Action", "Adventure"],
    "The Seven Deadly Sins": ["Action", "Adventure", "Fantasy"],
    "Akira": ["Action", "Sci-Fi", "Cyberpunk"],
    "Ghost in the Shell": ["Sci-Fi", "Cyberpunk"],
    "Princess Jellyfish": ["Comedy", "Romance"],
    "Wolf's Rain": ["Adventure", "Drama", "Fantasy"],
    "Death Note: The Last Name": ["Thriller", "Mystery", "Psychological"],
    "Cowboy Bebop: The Movie": ["Action", "Sci-Fi"],
    "Ghost in the Shell: Stand Alone Complex": ["Sci-Fi", "Cyberpunk"],
    "Spirited Away: The Cat Returns": ["Fantasy", "Adventure"],
    "Summer Wars": ["Sci-Fi", "Comedy", "Family"],
    "Grimgar: Ashes and Illusions": ["Adventure", "Fantasy", "Drama"],
    "The Rising of the Shield Hero": ["Action", "Adventure", "Fantasy"],
    "Attack on Titan: The Final Season": ["Action", "Fantasy", "Horror"],
    "Demon Slayer: Mugen Train": ["Action", "Fantasy", "Supernatural"],
    "Violet Evergarden": ["Drama", "Slice of Life"],
    "One Piece: Strong World": ["Action", "Adventure", "Comedy"],
    "Tokyo Ghoul:re": ["Action", "Horror", "Psychological"],
    "Anohana: The Flower We Saw That Day": ["Drama", "Slice of Life"],
    "Black Clover: Quartet Knights": ["Action", "Adventure", "Fantasy"],
    "Dragon Ball Z: Resurrection 'F'": ["Action", "Adventure"],
    "Attack on Titan: Lost Girls": ["Action", "Fantasy", "Mystery"],
    "Fruits Basket": ["Drama", "Romance", "Slice of Life"],
    "Code Geass: Lelouch of the Rebellion": ["Action", "Mecha", "Psychological"],
    "Clannad": ["Drama", "Romance", "Slice of Life"],
    "JoJo's Bizarre Adventure: Stardust Crusaders": ["Action", "Adventure", "Supernatural"],
    "Noragami": ["Action", "Supernatural"],
    "The Melancholy of Haruhi Suzumiya": ["Comedy", "Mystery", "Slice of Life"],
    "Haikyu!!": ["Comedy", "Sports"],
    "Sword Art Online": ["Action", "Adventure", "Fantasy"],
    "Re:Zero − Starting Life in Another World": ["Adventure", "Drama", "Fantasy"],
    "K-On!": ["Comedy", "Music", "Slice of Life"]
}

def recommend_anime(genres):
    """
    Recommends anime/movies based on the given genres.

    Args:
    - genres (list): A list of genres provided by the user.

    Returns:
    - list: A list of anime/movies that match the given genres.
    """
    matched_anime = []
    for anime, anime_genres in anime_database.items():
        if all(genre in anime_genres for genre in genres):
            matched_anime.append(anime)

    if not matched_anime:
        return "Sorry, we don't have any anime/movie matching those genres."

    return matched_anime

def main():
    print("Welcome to Anime Recommendation System!")
    print("Genres available: Action, Adventure, Fantasy, Horror, Mystery, Psychological, Romance, Drama, Thriller")

    while True:
        user_input = input("Enter your preferred genres (comma-separated, type 'quit' to exit): ").strip().title()

        if user_input.lower() == 'quit':
            print("Exiting...")
            break

        genres = [genre.strip() for genre in user_input.split(",")]
        recommendations = recommend_anime(genres)
        if isinstance(recommendations, list):
            print("Recommended anime/movies:")
            for anime in recommendations:
                print("-", anime)
        else:
            print("Recommended anime/movie:", recommendations)
        print()

if __name__ == "__main__":
    main()
```
